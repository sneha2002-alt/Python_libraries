```python
import requests
```

```python
URL = "https://stephen-king-api.onrender.com/api/books"
res = requests.get(URL)
```

```python
res
```

```python
json_data = res.json()
```

```python
import pandas as pd

df = pd.json_normalize(json_data["data"])

df = df[["id", "Title", "Year", "ISBN"]]
```

```python
df
```
