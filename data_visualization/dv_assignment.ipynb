```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
scores = np.random.normal(70, 10, 200)
plt.hist(scores, color='skyblue', edgecolor='black')
plt.ylabel("No. of students")
plt.xlabel("scores")
mean = np.mean(scores)
plt.axvline(
    x= mean,
    label = "Mean scores",
    color = "red",
    linewidth = 1.5,
    linestyle = "--"

)
```

```python
sns.set_theme(style = "darkgrid")
penguins = sns.load_dataset("penguins")
sns.relplot(data = penguins, x='flipper_length_mm', y='body_mass_g', hue='species')
```

```python
tips = sns.load_dataset("tips")
sns.boxplot(
    data = tips,
    x="day",
    y="total_bill",
    hue="sex"
)
plt.title("Total Bill Distribution by Day and Sex")
tips.groupby('day')['total_bill'].median().sort_values(ascending=False)
```

```python
sns.scatterplot(
    data = tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style = "smoker"
)
plt.title("Tip vs. Total Bill by Sex and Smoker Status")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
```

Smoker Analysis
Smokers do not clearly tip differently overall—both smoker/non-smoker points overlap heavily along the positive trend line (higher bills correlate with higher tips, ~15-20% rate). No distinct separation by smoker status emerges, though smokers slightly cluster at higher bills on weekends (e.g., Sat/Sun data). This suggests smoking habit has minimal impact on tip percentage after controlling for bill size.


```python
tips_pivot = tips.pivot_table(values="tip", index="day", columns="time", aggfunc="mean")
print(tips_pivot)
```

```python
sns.heatmap(tips_pivot,   cmap = "YlGnBu",
    annot = True,
    fmt=".2f")
plt.title("Average Tip by Day and Time")
plt.ylabel("Day")
plt.xlabel("Time")
```
